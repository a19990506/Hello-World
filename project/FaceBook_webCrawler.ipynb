```python
import selenium.webdriver.support.ui as ui
from time import sleep
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.webdriver.support import expected_conditions as EC
from selenium.webdriver.common.by import By
import csv
import pandas
```

```python
def _extract(id_name: str):
    """ Extracts the specified element by class name """
    return driver.find_element_by_id(id_name).text
```

```python
def scrape_job(url: str, wait: ui.WebDriverWait):
    """ Scrape the job info from the specified Url. A broswer driver MUST be initialized beforehand.
    :param url: the url of a detailed google job page.
    :param wait: contains timeout.
    :return: a dict wrapping all info.
    """
    
    # Wait until all required elements are generated.
    
    driver.get(url)
    
        # Wait until all required elements are generated.
    wait.until(lambda driver: EC.presence_of_element_located((By.CLASS_NAME, '_1kdc')))
    wait.until(lambda driver: EC.presence_of_element_located((By.CLASS_NAME, '_3-8r')))
    try:     
        sleep(2)
        # Extract job information.
        title = driver.find_element_by_class_name('_1kdc').text
        location = driver.find_element_by_class_name('_3-8r').text
        contents = driver.find_elements_by_class_name('_3-8q')[1:]
        try:
            responsibilities = contents[0].find_element_by_class_name('_6ad2').text
        except:
            responsibilities=""
        try:
            minimum_qual = contents[1].find_element_by_class_name('_6ad2').text
        except:
            minimum_qual=""
        try:
            preferred_qual = contents[2].find_element_by_class_name('_6ad2').text
        except:
            preferred_qual=""
            
        return {
            'title': title,
            'location': location,
            'minimum_qual': minimum_qual,
            'preferred_qual': preferred_qual,
            'responsibilities': responsibilities
        }
    except:
        print('noooooooooo')
        return None
    
    
```

```python
driver = webdriver.Chrome('chromedriver.exe')
driver.get('https://www.facebook.com/careers/jobs');
sleep(1)
driver.find_element_by_id('u_0_1q').send_keys('software')
driver.find_element_by_id('u_0_1q').send_keys(Keys.RETURN)
driver.implicitly_wait(10) 
# print(driver.page_source)  # 顯示全部html
wait = ui.WebDriverWait(driver,20, 0.5)
arr = []

while True:
    try:
        sleep(2)
        wait.until(lambda driver: EC.presence_of_element_located((By.CLASS_NAME,"_69jm")))
        ans =  driver.find_elements_by_class_name("_69jm")
        for a in range(0,len(ans),1):
            arr.append(ans[a].get_attribute('href'))
            
        wait.until(lambda driver:EC.presence_of_element_located((By.XPATH,"//a[text()='下一頁']")))
        next=driver.find_element_by_xpath("//a[text()='下一頁']")
        next.send_keys(Keys.RETURN)

    except:
        break
    
```

```python
arr[10]
```

```python
driver = webdriver.Chrome('chromedriver.exe')
jobs = []
for url in arr:
    data = scrape_job(url, wait)
    if data !="":
        jobs.append(data)
        print(len(jobs))
driver.quit()
```

```python
csv_file = 'facebook_jobs.csv'

with open(csv_file, 'w',newline='',encoding='utf-8') as f:
    w = csv.DictWriter(f, fieldnames = ['title', 'location', 'minimum_qual', 'preferred_qual', 'responsibilities'])
    w.writeheader()
    
    for job in jobs:
        if job is not None:
            w.writerow(job)
    
print('File written: ' + csv_file)
```
