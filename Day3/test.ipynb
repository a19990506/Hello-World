```python
import os
import jieba
from string import punctuation
add_punc='，。、【】“”：；（）《》‘’{}？！⑦()、%^>℃：.”“^-——=&#@￥ '
all_punc=punctuation+add_punc
```

```python
texts = os.listdir('./data3')
texts
```

```python
def sentence_cut(x):#cut words and delete punctuation
    x=re.sub(r'[A-Za-z0-9]|/d+','',x)#delet numbers and letters
    testline = jieba.cut(x,cut_all=False)
    testline=' '.join(testline)
    testline=testline.split(' ')
    te2=[]
    for i in testline:
        te2.append(i)
        if i in all_punc:
            te2.remove(i)
    return te2
```

```python
import re
corpus = []
for text in texts:
    with open('./data3/{}'.format(text), 'r',encoding = 'utf8') as file:
        corpus.append(file.read())
cut=list()
for a in range(0,33,1):
    cut.append(sentence_cut(corpus[a]))

for i in range(len(cut)):
    print(cut[i])
    
# m = 0.1

# for i in range(len(features)):
#     temp = textFrame.columns.values[[textFrame.iloc[i] >= m]]
#     cut.append(temp.tolist())
```

```python

import pandas as pd
from mlxtend.preprocessing import TransactionEncoder
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
te = TransactionEncoder()
te_ary = te.fit(cut).transform(cut)
df = pd.DataFrame(te_ary, columns=te.columns_)
graph = df.T
graph
```

```python
minsup=0.01
result = apriori(graph, min_support = minsup)
result
```

```python
rules = association_rules(result, metric="confidence", min_threshold=0.7)
rules.head()
```

```python

rules = association_rules(result, metric="lift", min_threshold=1.2)
rules.head()
```

```python
rules["antecedant_len"] = rules["antecedents"].apply(lambda x: len(x))
rules.tail()
```

```python

out = rules[ (rules["antecedant_len"] >= 2) &
       (rules['confidence'] > 0.4) &
       (rules['lift'] > 1.2) ]
out.head()
```
